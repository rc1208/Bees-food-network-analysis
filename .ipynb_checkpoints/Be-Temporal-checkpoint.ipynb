```python
def hist_graph(x,xlabel,ylabel,title,tau):
    
    x_min = min(x)
    x_max = max(x)
    bins = np.linspace (0, 120)
    
    plt.xlabel(xlabel, fontsize = 30)
    plt.ylabel(ylabel, fontsize = 30)
    plt.title(title, fontsize = 30)
    #plt.plot([0], 'k-', lw=2)
    plt.xticks(np.arange(0 , 120, 10),  rotation = 45)
    plt.grid(True)
    plt.hist(x, bins, alpha=0.75,label = str(tau))
   
    
    #plt.show()
```

```python
def plot_graph(ax,x,y,tau):
    
    #ax.set_xscale("log")
    #ax.set_yscale("log")
    ax.set_title("CCDF for different thresholds")
    ax.set_xlabel("log <k>")
    ax.set_ylabel("log Pr( k <=  <k>)")
    ax.scatter(x,y,label = str(tau))
    
    #plt.show()
```

### Rich Bees are more Bee-g Hearted!

```python
tau_list = [0,5,10,15,20,25,30,35] #for different tau's

food_set_dict = defaultdict(list)
count_of_food_all = data_day_one['time'].value_counts()
count_of_food_dict = defaultdict(int)
for tau in tau_list:
    
    G = nx.Graph()
    food_set = set()
    for index,row in data_day_one.iterrows():

            if row[5] >= tau:
                food_set.add(row[5])
                G.add_edge(row[1],row[2], weight = copy.deepcopy(row[5]))
    food_set_dict[tau] = food_set
   
    
    
    for item in food_set_dict[tau]:
        if item not in count_of_food_all:
            food_set_dict[tau].pop(item, None)
    
    count_of_food = {}
    for item in food_set_dict[tau]:
        if item in count_of_food_all:
            
            count_of_food[item] = count_of_food_all[item]  
    #print(count_of_food)
    count_of_food_dict[tau] = count_of_food
    #print(tau)
    
    sum_of_weights = sum(list(count_of_food_dict[tau].values()))
    
    x = list(count_of_food_dict[tau].keys())
    
    ccdf_food_dict = defaultdict(int)
    for key,val in count_of_food_dict[tau].items():
        s = 0
        list_greater = [i for i in x if key<=i]
        for j in list_greater:
            s += count_of_food[j]

        ccdf_food_dict[key] = s/sum_of_weights
    
    
    
    degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
    #print(degree_sequence)
    hist_graph(degree_sequence,'degree','count','Degree distribution for all thresholds',tau)
    #plot_graph(list(ccdf_food_dict.keys()),list(ccdf_food_dict.values()),tau)
plt.legend()
plt.show()
    
    

    
#print(count_of_food_dict)
```

```python

#fav_b_list = np.random.choice(list(data_day_3.id2.unique()),20, replace = False)
fav_b_list =  [1543,1671,1667,959,880,918,1484,1764,1966,1047]
#fav_b_list = [1671]
corr_list = []
for fav_b in fav_b_list:
    corr = 0
    fav_b_data_day_3 = data_day_3[data_day_3['id1'] == fav_b] 
    fav_b_data_day_4 = data_day_4[data_day_4['id1'] == fav_b]
    #print(fav_b_data_day_4)
    set_day_3 = set(fav_b_data_day_3.id2.unique())
    set_day_4 = set(fav_b_data_day_4.id2.unique())
    #print(set_day_3)
    num = len(set_day_3.intersection(set_day_4))
    #print(num)
    #print(len(set_day_3))
    corr = num/(math.sqrt(len(set_day_3)*len(set_day_4)))
    corr_list.append(corr)
    print("For"+str(fav_b)+ ",Correlation ="+ str(corr))
```

```python
def plot_graph(ax,x,y,tau):
    
    #ax.set_xscale("log")
    #ax.set_yscale("log")
    ax.set_title("Correlation for different bees")
    ax.set_xlabel("")
    ax.set_ylabel("log Pr( k <=  <k>)")
    ax.scatter(x,y,label = str(tau))
    
    #plt.show()
```

```python
from collections import 
fav_b_list =  [1543,1671,1667,959,880,918,1484,1764,1966,1047]
avg_corr_list = []
for day in range(3,12):
    day_first = data[data['day'] == day]
    day_second = data[data['day'] == (day+1)]
    corr_list = []
    print("day=",day)
    for fav_b in fav_b_list:
        corr = 0
        fav_b_data_day_first = day_first[day_first['id1'] == fav_b] 
        fav_b_data_day_second = day_second[day_second['id1'] == fav_b]
        #print(fav_b_data_day_4)
        set_day_first = set(fav_b_data_day_first.id2.unique())
        set_day_second = set(fav_b_data_day_second.id2.unique())
        #print(set_day_3)
        num = len(set_day_first.intersection(set_day_second))
        #print(num)
        #print(len(set_day_3))
        if len(set_day_first)*len(set_day_second) == 0:
            print("Dead bee = ",fav_b)
            continue
        else:
            corr = num/(math.sqrt(len(set_day_first)*len(set_day_second)))
            corr_list.append(corr)
        
        print("For "+str(fav_b)+ ",Correlation = "+ str(corr))
    print("Average correlation=", sum(corr_list)/10)
    avg_corr_list.append(sum(corr_list)/10)
print(avg_corr_list) 
```
