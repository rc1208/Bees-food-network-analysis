```python
import pandas as pd
import numpy as np
import networkx as nx
import sys
import matplotlib.pyplot as plt
import random
import time
#from nxsim import BaseNetworkAgent
from datetime import datetime
import collections
from collections import defaultdict
import copy
import math
nx.__version__
```

```python
# Set some parameters to apply to all plots. These can be overridden
# in each plot if desired
import matplotlib
# Plot size to 14" x 7"
matplotlib.rc('figure', figsize = (36, 18))
# Font size to 14
matplotlib.rc('font', size = 20)
# Add grid lines
matplotlib.rc('axes', grid = True)
# Set backgound color to white
matplotlib.rc('axes', facecolor = 'white')
```

```python
read_data = pd.read_csv("pnas.1713568115.sd01.txt") #read the data file, no cleaning required
read_data = read_data[read_data['trial'] == 1]  #take the data for the 1st trial only
data = read_data
data.head(10)
#print(len(data))
#type(data)



data['time'] = data['end'] - data['begin']#store the diff btw begin and end to get total time for each trans. in ms
data['time'] = data['time']/1000.0 #change 'ms' to 's'
data['time'] = data['time'].map(lambda x: [x]) #convert the time attr to list as as there are multiple iteractions between two bees
```

```python
data['begin'] = data['begin']/1000.0 #change 'ms' to 's'
data.head(10)
```

```python
data['UTC'] = data['begin'].map(lambda x: datetime.utcfromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
data.head(10)
```

```python
data['day'] = data['begin'].map(lambda x: int(datetime.utcfromtimestamp(x).strftime('%d')))
data['hours'] = data['begin'].map(lambda x: int(datetime.utcfromtimestamp(x).strftime('%H')))
data['min'] = data['begin'].map(lambda x: int(datetime.utcfromtimestamp(x).strftime('%M')))
data = data.sort_values('begin')
```

```python
data.head(20)
```

```python
days_list = data.day.unique()
days_list
```

### Network Stats(mean deg) for only Day 3 - without considering the edge weights

```python
data_day_3 = data[data['day'] == 3]
```

```python
data_day_3.head(20)
```

```python
hours_list = data_day_3.hours.unique()
hours_list
```

```python
def make_graph(data): #make a graph out of the dataframe structure
    G = nx.Graph() #initialize the graph
    for index, row in data.iterrows():
        #print(row[0],row[1],row[2])
        G.add_edge(row[0],row[1]) #add node1,node2,weight list iteratively to the graph
    
    return G
```

```python
def mean_degree(G): #to calculate the mean deg of a graph
    
    degrees = dict(G.degree())
    sum_of_edges = sum(degrees.values())
    mean_degree = sum_of_edges/len (degrees)
    return mean_degree
    
```

```python
hr_deg = defaultdict(int) #stores mean degree on a hourly basis for one day
for hr in hours_list:
    G = make_graph(data_day_3[data_day_3['hours'] == hr])
    hr_deg[hr] = mean_degree(G)
    #print(md)
```

```python
hr_deg
```

### Network Stats(mean deg) for all Days - without considering the edge weights - window is 1hr

```python
day_degree = defaultdict(dict)
for day in days_list:
    data_day = data[data['day'] == day] #data for that day
    #print(data_day)
    hours_list = data_day.hours.unique() #hours for that day
    hr_deg = defaultdict(int) #stores mean degree on a hourly basis for one day
    for hr in hours_list:
        G = make_graph(data_day[data_day['hours'] == hr])
        hr_deg[hr] = mean_degree(G)
        #print(md)

    #print(hr_deg)
    day_degree[day] = hr_deg

day_degree
```

```python
def plot_graph(d): #takes in dict, converts into list and then plots
    lists = sorted(d.items()) # sorted by key, return a list of tuples
    x, y = zip(*lists) # unpack a list of pairs into two tuples
    plt.plot(x, y)
```

```python
for day in days_list: #plot hourly deg value for each day
    plot_graph(day_degree[day]) #send dict of every day

plt.legend(days_list, loc='lower right')
plt.show()
```

### Network Stats(mean deg) for Day 3 - without considering the edge weights - window is 30 min

```python
data_day_3 = data[data['day'] == 3]

data_day_3.head(20)
```

```python
data[data['min']<15]
```

```python
hours_list = data_day_3.hours.unique()
hr_deg = defaultdict(int) #stores mean degree on a hourly basis for one day
half_hr_deg = defaultdict(int) #stores mean degree on a half hourly basis for one day
for hr in hours_list:
    data_hour = data_day_3[data_day_3['hours'] == hr]
    G = make_graph(data_hour)
    hr_deg[hr] = mean_degree(G)
    data_less_30 = data_hour[data_hour['min']<30]
    G1 = make_graph(data_less_30)
    half_hr_deg[hr] = mean_degree(G1)
    data_more_30 = data_hour[data_hour['min']>=30]
    G2 = make_graph(data_more_30)
    half_hr_deg[hr+0.5] = mean_degree(G2)
    #print(md)


plot_graph(hr_deg)
plot_graph(half_hr_deg)

plt.show()
```

```python
del half_hr_deg[5]
```

```python


plot_graph(hr_deg)
plot_graph(half_hr_deg)
plt.legend(['1hr time window', '30min time window'], loc='lower right')
plt.show()
```

```python
fifteen_min_deg = defaultdict(int) #stores mean degree on a half hourly basis for one day
for hr in hours_list:
    print(hr)
    data_hour = data_day_3[data_day_3['hours'] == hr]
    
    data_less_15 = data_hour[data_hour['min']<15]
    if(len(data_less_15) == 0):
        fifteen_min_deg[hr] = 0
    else:
        G1 = make_graph(data_less_15)
        fifteen_min_deg[hr] = mean_degree(G1)
    
    data_btw_15_and_30 = data_hour[(data_hour['min']>=15) & (data_hour['min']<30)]
    if(len(data_btw_15_and_30) == 0):
        fifteen_min_deg[hr+0.15] = 0
    else:
        G2 = make_graph(data_btw_15_and_30)
        fifteen_min_deg[hr+0.15] = mean_degree(G2)
    
    data_btw_30_and_45 = data_hour[(data_hour['min']>=30) & (data_hour['min']<45)]
    if(len(data_btw_30_and_45) == 0):
        fifteen_min_deg[hr+0.30] = 0
    else:
        G3 = make_graph(data_btw_30_and_45)
        fifteen_min_deg[hr+0.30] = mean_degree(G3)
    
    data_btw_45_and_60 = data_hour[(data_hour['min']>=45) & (data_hour['min']<60)]
    if(len(data_btw_45_and_60) == 0):
        fifteen_min_deg[hr+0.45] = 0
    else:
        G4 = make_graph(data_btw_45_and_60)
        fifteen_min_deg[hr+0.45] = mean_degree(G4)
```

```python

plot_graph(hr_deg)
plot_graph(half_hr_deg)
plot_graph(fifteen_min_deg)
plt.legend(['1hr time window', '30min time window','15min time window'], loc='lower right')
plt.show()
```
